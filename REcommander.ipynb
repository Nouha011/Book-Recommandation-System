# Preparing the data for vectorization
---------------------------------------------------------------------------------------------

```python
def clean_parentheses(author): #we will take out all the unnecessary brackets and their inside using regex
  return re.sub(r"\(.*\)", "", author)

df['author']=df['author'].apply(clean_parentheses)
df['author'][0:30]
```

```python
def standardize_authors(authors):
  #make the authors into a list
  author_list=authors.split(',')
  
  #standardize each name individually strip takes off beg and end spaces and replace in between
  author_list=[author.strip().replace(" ","").lower() for author in author_list]
  
  #join them back
  standardize_author = ' '.join(author_list)
  return standardize_author

df['author']=df['author'].apply(standardize_authors) #if we will re-run the code it will stick every thing together so this is done
df['author'][0:40]
```

```python
df.columns
```

```python
def clean_genres(genres):
    genres = genres.replace("[", "")
    genres = genres.replace("]", "")
    genres = genres.replace("'", "")

    genre_list = genres.split(",")
    genre_list = [genre.strip().replace(" ", "").lower() for genre in genre_list]
    unwanted_words=['lgbt', 'lesbian','gay','transgender', 'lesbianromance','gayromance','transgenderromance','lgbtromance'] #remove the things i don't want in my projects
    genre_list=[genre for genre in genre_list if genre not in unwanted_words]
    genres = " ".join(genre_list)
    return genres


df["genres"] = df["genres"].apply(clean_genres)
df["genres"][0:10]
```

```python
def lower_Everything(column):
  return column.str.lower()

df=df.apply(lower_Everything)
df.head(6)
```

```python
def clean_title(title):
  title = re.sub(r'[^A-Za-z\u0600-\u06FF\s]', '', title)
  title = re.sub(r'\s+', ' ', title)
  return title.strip()

df = df.dropna(subset=['title']) 
df['title']=df['title'].apply(clean_title)
df['title'][0:20]
```

```python
df['data'] = df[df.columns[0:]].apply(
    lambda x: ' '.join(x.astype(str)),
    axis=1
)
df.head(50)

df.to_csv('Cleaned_Books.csv',index=False)
```

# Vectorize-ing our data columns:
______________________________________________________________________________________________ 

```python
df2=pd.read_csv('Cleaned_Books.csv')
df2.head(6)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer=CountVectorizer() #create an instance of the function, we use it instead of the original function so we it can learn the vocabulary and we can use later
vectorized = vectorizer.fit_transform(df2['data']) # this is our matrix
vectorized
```

# Cosine Similarity 
_______________________________________________________________________________

```python
from sklearn.metrics.pairwise import cosine_similarity

similarities=cosine_similarity(vectorized) #nBooks*nBooks each cell(i,j) is the cosine similarity between the Book_i and Book_j
print(similarities)
```

```python
df_rec= pd.DataFrame(similarities,columns=df2['title'],index=df2['title']).reset_index() #we made both the columns and the rows(index) as the book title then we reset the index so we can make the rows 'normal rows' and get new column for indexes
df_rec.head(6)
#notice that the diagonal is always 1 because same books are 100% similar
```

# Recommending Books :
________________________________________________________________________________________________________________

```python
# Ask the user to type the book title
input_book = input("Enter the book title you want recommendations for: ")

# Check if the book exists in the dataset
if input_book not in df_rec.columns:
    print(f"Sorry, '{input_book}' was not found in the dataset.")
    print("Please make sure you enter the exact title (lowercase, no special characters).")
else:
    # Get top 11 most similar books (including the book itself)
    recommendations = pd.DataFrame(df_rec.nlargest(11, input_book)['title'])
    
    # Remove the input book from the recommendations
    recommendations = recommendations[recommendations['title'] != input_book]
    
    # Show the result
    print("\nTop 10 recommendations for '{}':\n".format(input_book))
    print(recommendations.to_string(index=False))
```
